# Setting Up

```python
import geopandas as gpd
```

```python
# CRS to be used
crs = 4269
```

```python
# open shapefiles
islands_polygon = gpd.read_file('files/2025-01-28_Puffin_colonies_shapefiles_FLT/islands_polygons.shp').to_crs(epsg=crs)
islands_polyline = gpd.read_file('files/2025-01-28_Puffin_colonies_shapefiles_FLT/islands_polyline.shp').to_crs(epsg=crs)
witless_bay_polyline = gpd.read_file('files/coastline/coastline.shp').to_crs(epsg=crs)
```

```python
# Display islands_polygon
islands_polygon
```

```python
# Display islands_polyline
islands_polyline
```

```python
# Display witless_bay_polyline
witless_bay_polyline
```

```python
# get GeoDataFrame of each island for each geometry
# Note that if iloc is used instead of geodataframe_x[col==val], a pandas series object is returned

# Gull island
gull_polygon = islands_polygon[islands_polyline.Island=='gull']
gull_polyline = islands_polyline[islands_polyline.Island=='gull']

# Great island
great_polygon = islands_polygon[islands_polyline.Island=='great']
great_polyline = islands_polyline[islands_polyline.Island=='great']
```
